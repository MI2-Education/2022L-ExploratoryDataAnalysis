Badanie zależności kolumn

```python
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(happinnes.iloc[3:13].corr(), annot=True, vmin=-1, vmax=1, cmap="BrBG")
plt.show()
```

Rozkład wartości koumny "Family"

```python
plt.hist(happinnes.Family, bins= 25)
plt.show()
```

Rozkład wartości kolumny 'Trust (Government Corruption)'

```python
plt.hist(happinnes['Trust (Government Corruption)'], bins= 25)
plt.show()
```

Rozkład wartości kolumny "Freedom"

```python
plt.hist(happinnes.Freedom, bins= 25)
plt.show()
```

Zależność między 'Family' a 'Happiness Score'

```python
sns.scatterplot(data = happinnes, y = 'Happiness Score', x = 'Family', hue = 'Region')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

Im wyższa jest wartośc Family to tym wyższy wynik 'Happiness Score'

Zależność miedzy 'Happiness Score' a 'Happiness Score'

```python
sns.scatterplot(data = happinnes, x = 'Happiness Score', y = 'Trust (Government Corruption)', hue = 'Region')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

W pewnych regionach im większe wartości 'Trust' tym większe wartości 'Happiness Score'
