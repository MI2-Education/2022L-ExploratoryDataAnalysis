# Opis zbioru danych

Ramka danych będąca przedmiotem badań w mojej analizie zawiera następujące kolumny:
1. "Overall rank" - miejsce danego kraju w rankingu bazującym na 'Happiness Score' - poziomie szczęścia
2. "Country or region" - kraj lub wyróżniony region
3. "Score" - poziom szczęścia mierzony na podstawie drabiny Cantril. Uczestnicy ankiety oceniali swoje szczeście w skali od 0 do 10, mając za zadanie wyobrazić sobie drabinę, a następnie uplasować swoje życie na odpowiednim jej szczeblu, gdzie "najlepsze" życie jest na 10 szczeblu, a "najgorsze" na szczeblu 0. 
4. "GDP per capita" - stopień w jakim PKB przyczynia się do obliczenia wskaźnika zadowolenia.
5. "Social support" - stopień w jakim wsparcie społeczeństwa przyczynia się do obliczenia wskaźnika zadowolenia.
6. "Healthy life expectancy" - stopień w jakim długość życia przyczynia się do obliczenia wskaźnika zadowolenia.
7. "Freedom to make life choices" - stopień w jakim wolność przyczynia się do obliczenia wskaźnika zadowolenia.
8. "Generosity" - stopień w jakim hojność przyczynia się do obliczenia wskaźnika zadowolenia.
9. "Perceptions of corruption" - stopień w jakim korupcja przyczynia się do obliczenia wskaźnika zadowolenia.
Liczba wierszy ramki danych wynosi 156 - wiersze odnoszą się do krai, które brały udział w badaniu, są to unikalne wartości.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('2019.csv')
df.head()
```

# Rozkład zmiennych

```python
plt.hist(x = df['Score'], edgecolor = "black")
plt.xlabel('Happiness score')
plt.ylabel('Ilość krai')
plt.title('Rozkład poziomu szczęścia na świecie')
```

Na podstawie powyższego histogramu możemy zauważyć, że poziom szczęścia państw na świecie, znajduje się zazwyczaj między 4.5 a 6.5 punktów w skali Happiness score.

Przyjrzymy się teraz rozkładowi zmiennych opisujących wpływ na wynik poziomu szczęścia względem poszczególnych kategorii.

```python
plt.hist(x = df['GDP per capita'], edgecolor = "black")
plt.xlabel('PKB na mieszkańca')
plt.ylabel('Ilość krai')
plt.title('Rozkład wpływu PKB na poziom szczęścia')
```

Z powyższgo histogramu mozemy odczytać, że PKB ma zmienny wpływ na poziom szczęścia ludzi w danym kraju.

```python
plt.hist(x = df['Social support'], edgecolor = "black")
plt.xlabel('Wsparcie społeczne')
plt.ylabel('Ilość krai')
plt.title('Rozkład wpływu wsparcia społecznego na poziom szczęścia')
```

```python
plt.hist(x = df['Healthy life expectancy'], edgecolor = "black")
plt.xlabel('Długość zycia')
plt.ylabel('Ilość krai')
plt.title('Rozkład wpływu długości życia na poziom szczęścia')
```

```python
plt.hist(x = df['Freedom to make life choices'], edgecolor = "black")
plt.xlabel('Wolność wyboru')
plt.ylabel('Ilość krai')
plt.title('Rozkład wpływu wolności wyboru na poziom szczęścia')
```

```python
plt.hist(x = df["Generosity"], edgecolor = "black")
plt.xlabel('Hojność')
plt.ylabel('Ilość krai')
plt.title('Rozkład wpływu hojności na poziom szczęścia')
```

```python
plt.hist(x = df["Perceptions of corruption"], edgecolor = "black")
plt.xlabel('Odczuwanie korupcji w kraju')
plt.ylabel('Ilość krai')
plt.title('Rozkład wpływu korupcji na poziom szczęścia')
```

### Wnioski:
Jeśli za kryterium rozróżniania czy dana zmienna ma wpływ na wartość poziomu szczęścia czy nie przyjmiemy wartość 0.8 (tj. jeśli dana zmienna ma wartość większą niż 0.8, to ma wpływ znaczący na poziom szczęścia), to możemy zauważyć, że:
* w większości krai znaczący wpływ na wartość poziomu szczęścia ma PKB na mieskzańca, długość życia oraz wsparcie społeczne
* zmienne opisujące wolność wyboru, hojność i odczuwanie korupcji nie mają aż tak dużego wpływu na wartość poziomu szczęścia
* zmienną mającą najmniejszy wpływ na wartość poziomu szczęścia jest odczuwanie korupcji

Ponadto większość krai ma szczęście na poziomie 4.5-6.5.

# Zależności w zbiorze danych

Ponieważ na podstawie analizy poszczególnych zmiennych okazało się, że największy wpływ na poziom szczęścia mają PKB na jednego mieszkańca oraz wsparcie społeczne, postanowiłam zbadać zależności właśnie między tymi zmiennymi oraz między nimi a poziomem szczęścia. Oto wyniki:

```python
sns.scatterplot(data=df, x='GDP per capita', y='Score', alpha = 0.4)
plt.xlabel('PKB na mieszkańca')
plt.ylabel('Wartość poziomu szczęścia')
plt.title("Korelacja między wpływem PKB na poziom szczęścia a warością poziomu szczęścia")
plt.show()
```

```python
sns.scatterplot(data=df, x='Social support', y='Score', alpha = 0.4)
plt.xlabel('Wsaprcie społeczne')
plt.ylabel('Wartość poziomu szczęścia')
plt.title("Korelacja między wpływem wsparcia społecznego na poziom szczęścia a warością poziomu szczęścia")
plt.show()
```

```python
sns.scatterplot(data=df, x='Social support', y='GDP per capita', alpha = 0.4)
plt.xlabel('Wsaprcie społeczne')
plt.ylabel('PKB na mieszkańca')
plt.title("Korelacja między wpływem wsparcia społecznego a wpływem PKB na poziom szczęścia")
plt.show()
```

### Wnioski:
Na podstawie powyższych wykresów możemy zauważyć, że:
* nie widać istotnej korelacji między wpływem PKB na jednego mieszkańca na poziom szczęścia a samymy poziomem szczęścia
* im wyższy jest wpływ wsparcia społecznego na wartość poziomu szczęścia tym wyższa jest wartość poziomu szczęścia
* wraz ze wzrostem wpływu wsparcia społecznego na wartość poziomu szczęścia rośnie także wpływ PKB na jednego mieszkańca na poziom szczęścia

# Wnioski końcowe
Na podstawie powyższych analiz, można zauważyć, że większość państw świata ma poziom szczęścia mieszczący się między 4.5-6.5. Patrząc na to, że poziom szczęścia jest oceniany w skali od 0-10 oznacza to, że w większości krai na świecie ludzie są średnio zadowoleni ze swojego życia. 

Kryteriami, które mają największy wpływ na poziom szczęścia są PKB na jednego mieszkańca, wsparcie społeczne oraz długość życia. Ponadto wiele osób które uważa, że wsparcie społeczne ma wpływ na poziom ich szczęścia żyje w kraju o wyższym poziomie szczęścia, co może oznaczać, że wsparcie społeczne, ma rzeczywiście duży wpływ na poziom zadowolenia ludzi, oraz, że zaczynamy to dostrzegać dopiero wtedy, gdy rzeczywiście takiego wsparcia doświadczamy. Można zauważyć także, że w społeczności dla której wsparcie społeczne jest istotnym kryterium dla oceny poziomu szczęścia, istotnym kryterium jest także wartość PKB na jednego mieszkańca. Jednak dla wpływu wartości PKB na jednego mieszkańaca na poziom szczęścia nie możemy zauważyć korelacji z samym poziomem.
